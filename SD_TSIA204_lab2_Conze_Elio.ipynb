```python
import numpy as np
from numpy import linalg
import scipy
import statsmodels.api as sm
import statsmodels

n = len(df.index)


#ols_k is a linear regression between Xk and Y with no intercept
#returns the linear regression coefficients
#I used the statsmodel ols (only here) because it dealt more easily with DataFrames
def ols_k(Y,Xk):
  ols = sm.OLS(Y,Xk)
  olsResults = ols.fit()
  return olsResults.params

#Gram Matrix of Xk
def G_k(Xk):
  data = {'1' : [1 for i in range(len(Xk))], 'Xk' : Xk}
  one_Xk = pd.DataFrame(data=data, index=Xk.index)
  one_Xk_transposed = one_Xk.transpose()
  gram = one_Xk_transposed.dot(one_Xk)
  return gram.mul(1/n)

#s_n(Xk) as defined in the course
def s_k(Xk):
  gram = G_k(Xk)
  return (linalg.inv(gram)[1][1])**0.5

#sigma_n(Y,Xk) as defined in the course
def sigma_k(Y,Xk):
  olsk = ols_k(Y,Xk)
  result_dataframe = Y.sub(Xk.mul(olsk[0]))
  result_matrix = result_dataframe.to_numpy()
  norm = np.linalg.norm(result_matrix)
  squared_norm = norm**2
  return (squared_norm/(n-2))**0.5


#T_n(Y,Xk) almost as defined in the course :
#the ols result is in absolute values
#and the test defined in the course is multiplied by sqrt(n)
def test_k(Y,Xk):
  olsk = ols_k(Y,Xk)
  sk = s_k(Xk)
  sigmak = sigma_k(Y,Xk)
  return (n**0.5)*abs(olsk.mul(1/(sk * sigmak)))


#argmax of a list that can contain None values
def argmax(list):
  l = len(list)
  index = 0
  for i in range(l):
    if list[i]!=None:
      index = i
  for k in range(l):
    if list[k]!=None:
      if (list[k]>list[index]):
        index = k
  return index

#forward variable selection algorithm, with a slight difference as in the course
#instead of "r = r - (1n, Xk*) * Theta_n(Y,Xk*)"
#it is "r = r - (1n, Xk*) * Theta_n(r,Xk*)"
def forward_variable_selection(Y,X):
  r=Y
  S=[]
  pvalues = []
  p = len(X.columns)
  A=[i for i in range(p)]
  for i in range(p):
    test=[None for j in range(p)]
    for k in A:
      if k not in S:
        Xk = X[X.columns[k]]
        testk = test_k(r,Xk)
        test[k] = testk.iloc[0]
    k_star = argmax(test)
    S.append(k_star)
    pvalues.append(2*(1-scipy.stats.norm.cdf(test[k_star])))
    r = r.sub(X[X.columns[k_star]].mul(ols_k(r,X[X.columns[k_star]])[0]))
  final = []
  for k in range(len(S)):
    final.append([S[k],pvalues[k]])
  final_dataframe1 = pd.DataFrame(data = final, columns = ["k","pvalue"])
  final_dataframe = pd.DataFrame(data = final_dataframe1["pvalue"], index = final_dataframe1["k"])
  return final


ordered_variables = forward_variable_selection(std_data_train["fat"],std_data_train.loc[:, std_data_train.columns != 'fat'])
```

```python
from sklearn.linear_model import LinearRegression

from sklearn.model_selection import KFold


#function fold folds the X,Y sets into K folds
#returns 2 lists of dataframe :
#1 list of K dataframes representing X, 1 list of K dataframes representing Y
def fold(X,Y, K):
  n=len(X)
  L=[i for i in range(n)]
  divider = n//K
  splitlistX = []
  splitlistY = []
  for i in range(K-1):
    splitlistX.append(L[i*divider:(i+1)*divider])
    splitlistY.append(L[i*divider:(i+1)*divider])
  splitlistX.append(L[(K-1)*divider:])
  splitlistY.append(L[(K-1)*divider:])
  for iterator in range(len(splitlistX)):
    for i in range(len(splitlistX[iterator])):
      splitlistX[iterator][i] = X.iloc[[splitlistX[iterator][i]]]
      splitlistY[iterator][i] = Y.iloc[[splitlistY[iterator][i]]]

  split_dflistX = []
  split_dflistY = []
  for iterator in range(len(splitlistX)):
    columnsX = [X.columns[i] for i in range(len(X.columns))]
    indexX= [splitlistX[iterator][i].index[0] for i in range(len(splitlistX[iterator]))]
    dataX = [splitlistX[iterator][i].iloc[0] for i in range(len(splitlistX[iterator]))]
    tempX = pd.DataFrame(data = dataX, columns = columnsX, index= indexX)
    split_dflistX.append(tempX)

    columnsY = [Y.columns[i] for i in range(len(Y.columns))]
    indexY= [splitlistY[iterator][i].index[0] for i in range(len(splitlistY[iterator]))]
    dataY = [splitlistY[iterator][i].iloc[0] for i in range(len(splitlistY[iterator]))]
    tempY = pd.DataFrame(data = dataY, columns = columnsY, index= indexY)
    split_dflistY.append(tempY)
  return split_dflistX, split_dflistY


#auxiliary function that fusions a list of dataframes into a single dataframe
def fusion(df_list):
  df0 = df_list[0]
  for i in range(1,len(df_list)):
    df0 = df0.append(df_list[i])
  return df0

#cross validation function
#returns the score of the cross validation
def cross_validation(model,X,Y,K):
  splitX, splitY = fold(X,Y,K)
  scores = []
  for i in range(K):
    splitXtrainlist = [splitX[k] for k in range(len(splitX)) if k!=i]
    splitYtrainlist = [splitY[k] for k in range(len(splitY)) if k!=i]
    Xtrain = fusion(splitXtrainlist)
    Ytrain = fusion(splitYtrainlist)
    Xtest = splitX[i]
    Ytest = splitY[i]
    model.fit(Xtrain,Ytrain)
    r2 = model.score(Xtest,Ytest)
    scores.append(r2)
  return np.mean(scores)

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

Xdf = df.loc[:, df.columns != 'fat']
Ydf = df[["fat"]]

scaler = StandardScaler()
X_array = scaler.fit_transform(Xdf)
Y_array = scaler.fit_transform(Ydf)
std_Xdf = pd.DataFrame(data = X_array, index = Xdf.index, columns = Xdf.columns)
std_Ydf = pd.DataFrame(data = Y_array, index = Ydf.index, columns = Ydf.columns)

#OLS cross validation

print("Cross validation for the OLS model :",cross_validation(LinearRegression(fit_intercept=False), std_Xdf, std_Ydf, 5))
```

The results of the cross validation for the OLS are good.

We then generate bootstrap samples.

```python
from scipy import stats

#creates a sample from X and Y
#returns a sample of X and a sample of Y with same indexes
def bootstrap_sample(X, Y):
  k = random.randint(0, len(X)-1)
  df0 = X.iloc[[k]]
  df1 = Y.iloc[[k]]
  for i in range(1,len(X)):
    k = random.randint(0, len(X)-1)
    row = X.iloc[[k]]
    row2 = Y.iloc[[k]]
    df0 = df0.append(row)
    df1 = df1.append(row2)
  return df0, df1

#creates a set of samples
def bootstrap_samples(X,Y, nb_samples):
  L=[]
  L2=[]
  for i in range(nb_samples):
    L.append(bootstrap_sample(X,Y)[0])
    L2.append(bootstrap_sample(X,Y)[1])
  return L,L2

def bootstrap_samples_regression(X,Y,nb_samples):
  bootstrap_samples_X_list, bootstrap_samples_Y_list = bootstrap_samples(X,Y,nb_samples)
  L=[]
  for i in range(len(bootstrap_samples_X_list)):
    L.append(LinearRegression(fit_intercept=False).fit(bootstrap_samples_X_list[i], bootstrap_samples_Y_list[i]).coef_[0][0])
  return L

#gives the upper and lower value of the prediction interval of level 1-alpha at point x
def pred_interval(x,X,Y,Ypred,theta, alpha):
  resid = Y.to_numpy() - Ypred
  sigma2_est = np.sum(resid**2) / (len(X) - 2)
  X_array = X.to_numpy()
  term = np.dot(np.transpose(X_array),X_array)
  term2 = term[0,0]
  term2 = np.sqrt(1 + x**2/term2)
  term2 = stats.t.ppf(q = 1 - alpha / 2, df = len(X)-2) * sigma2_est * term2
  upper_value = x*theta + term2
  lower_value = x*theta - term2
  return lower_value, upper_value

bootstrap_thetas = bootstrap_samples_regression(v40, df[["fat"]], 50)

abscisses = np.linspace(v40.min(),v40.max(), 1000)

plt.figure(figsize = (16,4))
confidence_alpha=0.2

for theta in bootstrap_thetas:
  plt.plot(abscisses, f(abscisses, theta), 'b', alpha=0.4)
  plt.plot(abscisses, pred_interval(abscisses, v40, df[["fat"]], f(v40.to_numpy(), theta), theta, alpha = confidence_alpha)[0], 'r', alpha=0.4)
  plt.plot(abscisses, pred_interval(abscisses, v40, df[["fat"]], f(v40.to_numpy(), theta), theta, alpha = confidence_alpha)[1], 'r', alpha=0.4)


print("Prediction interval of level 1-alpha = ", (1-confidence_alpha)*100, "%")

plt.legend(('Regression line', 'Lower bound of prediction interval', 'Upper bound of prediction interval'))
plt.xlabel("x")
plt.ylabel("y")
plt.title("Regression model & prediction interval for each bootstrap sample")
plt.axis("tight")
plt.show()

```

##c)

The quantity of interet to know if a PC is important is the eigenvalue of the covariance matrix that is tied with the PC (same column).


We have the relation $\lambda_i = \frac{s_i^2}{n-1}$ with $s_i$ the singular value of X at column i.

The greater the eigenvalue is, the more the PC has an impact on the variance.
